**Type Casting**

```python
 arr=np.array([1,2,3.0],dtype="int64")
 print(arr)

 b=np.array([1,2,3.0],dtype="float64")
 print(b)

 new_b=b.astype("int64")
 print(new_b)
```

```python
# Reshaping Arrays

d=np.array([[3,5],[4,6]])
print(d)

d.ndim #check dimension
d.shape #check shape
d.dtype
d.reshape(1,4)



# Flatten - -> convert array into 1D array without returning a view

d.flatten()
```
